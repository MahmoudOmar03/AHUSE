```python
import os
import cv2

train_img_dir = path + '/css-data/train/images'
valid_img_dir = path + '/css-data/valid/images'

image_dirs = [train_img_dir, valid_img_dir]

for img_dir in image_dirs:
    for filename in os.listdir(img_dir):
        img_path = os.path.join(img_dir, filename)
        img = cv2.imread(img_path)

        if img is not None:
            h, w, _ = img.shape
            if h != 640 or w != 640:
                print(f"Image '{filename}' in '{img_dir}' is not 640x640. Size: {w}x{h}")
        else:
            print(f"Could not read image: {filename} in {img_dir}")
```

```python
import yaml

# Define the paths to your dataset
train_img_path = path + '/css-data/train/images'
val_img_path = path + '/css-data/valid/images'
train_label_path = path + '/css-data/train/labels'
val_label_path = path + '/css-data/valid/labels'

# Define the class names
classes = ['hardhat', 'mask', 'no-hardhat', 'no-mask', 'no-safety-vest', 'person', 'safety-vest']

# Create the data dictionary
data = {
    'train': train_img_path,
    'val': val_img_path,
    'nc': len(classes),
    'names': classes
}

# Write the dictionary to a YAML file
with open('data.yaml', 'w') as f:
    yaml.dump(data, f)

print("data.yaml file created successfully!")
```

```python
from ultralytics import YOLO

model=YOLO('yolov8n.pt')
```

```python
model.train(data=os.path.abspath('data.yaml'), epochs=170, imgsz=640)
```
